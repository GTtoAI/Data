```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset and convert it into a pandas DataFrame
iris = load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
iris_df['target'] = iris_df['target'].astype(int)
```

```python
# Task 3: Line Plot
plt.figure(figsize=(12, 6))
for feature in iris['feature_names']:
    plt.plot(iris_df[feature][iris_df['target'] == 0], label=f'Setosa - {feature}', marker='o')
    plt.plot(iris_df[feature][iris_df['target'] == 1], label=f'Versicolor - {feature}', marker='o')
    plt.plot(iris_df[feature][iris_df['target'] == 2], label=f'Virginica - {feature}', marker='o')

plt.title('Line Plot for Each Feature Against Species')
plt.xlabel('Samples')
plt.ylabel('Feature Value')
plt.legend()
plt.show()
```

```python

# Task 4: Scatter Plot
plt.figure(figsize=(15, 10))
for i in range(3):
    for j in range(i+1, 4):
        plt.scatter(iris_df.iloc[:, i], iris_df.iloc[:, j], c=iris_df['target'], cmap='viridis', label=iris['target_names'])
        plt.xlabel(iris['feature_names'][i])
        plt.ylabel(iris['feature_names'][j])
        plt.title(f'Scatter Plot for {iris["feature_names"][i]} and {iris["feature_names"][j]} Colored by Species')
        plt.legend()
        plt.show()
```

```python
# Task 5: Histogram
plt.figure(figsize=(15, 10))
for i, feature in enumerate(iris['feature_names']):
    plt.hist([iris_df[feature][iris_df['target'] == 0], iris_df[feature][iris_df['target'] == 1], iris_df[feature][iris_df['target'] == 2]],
             bins=10, alpha=0.5, label=iris['target_names'])
    plt.title(f'Histogram for {feature} Colored by Species')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
# Task 6: Bar Chart
species_counts = iris_df['target'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(species_counts.index, species_counts.values, color=['blue', 'orange', 'green'])
plt.title('Bar Chart of Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(species_counts.index, labels=iris['target_names'])
plt.show()
```

```python
# Task 7: Heatmap
correlation_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.title('Heatmap of Correlation Matrix')
plt.xticks(np.arange(4), labels=iris['feature_names'])
plt.yticks(np.arange(4), labels=iris['feature_names'])
plt.show()
```
